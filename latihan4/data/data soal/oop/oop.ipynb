```python
import re
import pandas as pd
```

```python
data = pd.read_csv('oop.csv')
data['Full Code'] = data['Output']
data.head()
```

```python
def add_placeholder(code):
    """
    Gantikan bagian tertentu dalam kode dengan placeholder '___'.
    """
    # Pisahkan kode menjadi baris-baris
    lines = code.split('\n')

    # Pola regex yang akan dicari
    patterns = [
        r'def\s+\w+\(.*\):',  # Pola untuk mendeteksi definisi fungsi
        r'if\s+.*:',          # Pola untuk kondisi if
        r'for\s+.*\s+in\s+.*:',  # Pola untuk loop for
        r'while\s+.*:'        # Pola untuk loop while
    ]

    # Ganti baris yang cocok dengan pola menjadi '___'
    for i, line in enumerate(lines):
        for pattern in patterns:
            if re.search(pattern, line):
                lines[i] = "    ___"  # Placeholder
    return "\n".join(lines)
```

```python
# Terapkan fungsi pada kolom 'Full Code'
data['Template'] = data['Full Code'].apply(add_placeholder)
print(data['Template'][10])
```

```python
data.head()
```

```python
# Simpan dataset yang telah diproses
data.to_csv('function_with_placeholders.csv', index=False)
```

```python
def create_placeholder_and_answer(code):
    """
    Membuat placeholder untuk kode dan mengembalikan tuple (template, answer).
    """
    # Pisahkan kode menjadi baris-baris
    lines = code.split('\n')

    # Pola regex untuk mengganti baris tertentu dengan placeholder
    patterns = [
        r'def\s+\w+\(.*\):',  # Pola untuk mendeteksi definisi fungsi
        r'if\s+.*:',          # Pola untuk kondisi if
        r'for\s+.*\s+in\s+.*:',  # Pola untuk loop for
        r'while\s+.*:'        # Pola untuk loop while
    ]

    # Ganti baris yang cocok dengan pola menjadi placeholder
    template_lines = []
    for line in lines:
        line_modified = line  # Simpan baris asli
        for pattern in patterns:
            if re.search(pattern, line):
                line_modified = "    ___"  # Gantikan baris dengan placeholder
                break
        template_lines.append(line_modified)

    # Gabungkan kembali template dengan placeholder
    template_code = "\n".join(template_lines)

    # Jawaban asli tetap sama
    answer_code = code

    return template_code, answer_code
```

```python
# Terapkan fungsi untuk membuat template dan answer
data[['Template', 'Answer']] = data['Full Code'].apply(
    lambda x: pd.Series(create_placeholder_and_answer(x))
)
```

```python
# Simpan dataset baru ke file CSV
data.to_csv('function_with_placeholders_and_answers.csv', index=False)
```
